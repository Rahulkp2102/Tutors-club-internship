```python
from pptx import Presentation
from pptx.util import Inches
from pptx.enum.text import PP_ALIGN
from pptx.dml.color import RGBColor
from pptx.enum.dml import MSO_THEME_COLOR
from pptx.util import Pt


# Open the existing PowerPoint presentation
prs = Presentation('STUDENT PERFORMANCE REPORT.pptx')

# Add a slide
slide1 = prs.slides.add_slide(prs.slide_layouts[0])


# Set the background image
img_path = r"G:\Tutors club internship\B.jpg"
pic = slide1.shapes.add_picture(img_path, left=0, top=0, width=prs.slide_width, height=prs.slide_height)

# Add a centered heading
tf = slide1.shapes.add_textbox(left=0, top=Inches(3), width=prs.slide_width, height=prs.slide_height).text_frame

tf.text = "STUDENT PERFORMANCE REPORT"
tf.paragraphs[0].font.bold = True
tf.paragraphs[0].font.size = Pt(50)
tf.paragraphs[0].font.color.rgb = RGBColor(0, 0, 0)
tf.paragraphs[0].alignment = PP_ALIGN.CENTER

# Save the PowerPoint presentation
prs.save("STUDENT PERFORMANCE REPORT.pptx")
```

```python
# definig a function for all the graph slides
def graph_slide(image_path):
    slide = prs.slides.add_slide(prs.slide_layouts[0])
    background_image =  r"G:\Tutors club internship\B.jpg"
    pic = slide.shapes.add_picture(background_image, left=0, top=0, width=prs.slide_width, height=prs.slide_height)
    graph_image = image_path
    pic = slide.shapes.add_picture(graph_image, left=Inches(1.5), top=Inches(1), width=Inches(10), height=Inches(6))
    
    
graph_slide(r"G:\Tutors club internship\english.png")
# Save the PowerPoint presentation
prs.save("STUDENT PERFORMANCE REPORT.pptx")
```
