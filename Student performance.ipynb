```python
def bar_graph_each_suject(row_index,subject):
    # select the specified row from the dataframe
    row_data = data.loc[row_index]

    # convert the selected row into a new dataframe
    subject_data = pd.DataFrame(row_data).T

    x=["JUNE","JULY","AUGUST"]
    w=0.4
    # adding columns needed to plot into array 
    test=subject_data[['TEST 1(50)','TEST 2(50)','TEST 3(50)']].values
    Test = test.flatten()
    retest=subject_data[['RETEST 1(50)','RETEST 2(50)','RETEST 3(50)']].values
    Retest = retest.flatten()

    bar1= np.arange(len(x))
    bar2= [i+w for i in bar1]
    
    # plotting graph
    plt.figure(figsize=(12,6))
    plt.style.use('ggplot')
    plt.bar(bar1, Test, w, color='#808000', label='test')
    plt.bar(bar2, Retest, w, color='#006400', label='retest')

    plt.title(subject,fontsize=20, color='black', fontweight='bold')
    plt.xlabel('Months')
    plt.ylabel('score')
    plt.ylim(30, 50)

    plt.xticks(bar1+w/2,x)
    plt.legend()
    plt.show()
# call the function
bar_graph_each_suject(row_index=1,subject='English')
```

```python
# managing the outliers
c= data.select_dtypes(exclude=['object'])
for i in c:
    q1=np.percentile(data[i],25,interpolation='midpoint')
    q2=np.percentile(data[i],50,interpolation='midpoint')
    q3=np.percentile(data[i],75,interpolation='midpoint')
    iqr=q3-q1
    up_lim=q3+1.5*iqr
    low_lim=q1-1.5*iqr
    print('\033[1;4m','{}'.format(i),'\033[0m')
    print('Q1:',q1)
    print('Q2:',q2)
    print('Q3:',q3)
    print('IQR:',iqr)
    print('lower limit:',low_lim)
    print('upper limit:',up_lim)
    outlier=[]
    for x in data[i]:
        if ((x<low_lim) or (x>up_lim)):
            outlier.append(x)
    
    data[i]=np.where(data[i]<low_lim,low_lim,np.where(data[i]>up_lim,up_lim,data[i]))
    
```

```python
def table_for_each_subject(row_index):
    # select the specified row from the dataframe
    row_data = data.loc[row_index]
    # convert the selected row into a new dataframe
    subject_data = pd.DataFrame(row_data).T
    subject_data

    # creating table for each subject to present in ppt
    table = pd.DataFrame(columns=["sub","test","retest","improvement"])
    for i in range(3):
        table = table.append({}, ignore_index=True)
    
    
    table.iloc[0]= subject_data[['SUB','TEST 1(50)', 'RETEST 1(50)','%IMPROVEMENT ']].values
    table.iloc[1] = subject_data[['SUB','TEST 2(50)', 'RETEST 2(50)', '%IMPROVEMENT .1']].values
    table.iloc[2] = subject_data[['SUB','TEST 3(50)', 'RETEST 3(50)', '%IMPROVEMENT .2']].values
    table= table.assign(months = ['June','July','August'])
    average = table['retest'].mean()
    table.loc['average','retest']=average
    return table
```

```python
def performance_analysis(data,subject):
    
    fig, ax = plt.subplots()
    table = ax.table(cellText=data.values, colLabels=data.columns, loc='center')
    table.auto_set_font_size(False)
    table.set_fontsize(14)
    table.scale(2.5, 2)

    plt.axis('off')
    plt.title('PERFORMANCE ANALYSIS '+subject,fontweight="bold", fontstyle="oblique")
    return plt.savefig(subject+'.png', bbox_inches='tight')
```

```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder

# Create a label encoder for the first categorical column
le1 = LabelEncoder()
data['SUB'] = le1.fit_transform(data['SUB'])

# Create a label encoder for the second categorical column
le2 = LabelEncoder()
data['TUTOR'] = le2.fit_transform(data['TUTOR'])
```

# Modeling

```python
#Model Building
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error as mse
from sklearn.metrics import r2_score as r2

from sklearn.linear_model import Ridge, Lasso, LinearRegression
from sklearn.neural_network import MLPRegressor
from sklearn.ensemble import RandomForestRegressor
from catboost import CatBoostRegressor
from xgboost import XGBRegressor,XGBRFRegressor
from lightgbm import LGBMRegressor
from sklearn.tree import DecisionTreeRegressor
```

```python
X=data.drop('TERM I MODEL EXAM(100)', axis=1)
y=data['TERM I MODEL EXAM(100)']
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,random_state=42,test_size=0.2)
X_train.shape
```

```python
models = {
    'Linear Regression(Ridge)' : Ridge(),
    '          Neural Network' : MLPRegressor(),
    '           Random Forest' : RandomForestRegressor(),
    '            DecisionTree' : DecisionTreeRegressor(),
    '                Catboost' : CatBoostRegressor(verbose=0),
    '       XGBoost Regressor' : XGBRegressor(),
    '    XGBoost RF Regressor' : XGBRFRegressor(),
    'Linear Regression(lasso)' : Lasso(),
    '       Linear Regression' : LinearRegression(),
    ' Light Gradient Boosting' : LGBMRegressor()
}

for name, model in models.items():
    model.fit(X_train, y_train)
    print(name + ' trained.')
```

```python
def get_rmse(y_true,y_pred):
    rmse = np.sqrt(np.mean((y_test-y_pred)**2))
    return rmse

def get_r2(y_true,y_pred):
    r2 = 1 - (np.sum((y_test - y_pred)**2) / np.sum((y_test - y_test.mean())**2))
    return r2
```

```python
for name, model in models.items():
    y_pred = model.predict(X_test)
    rmse = get_rmse(y_test, y_pred)
    print(name + ' RMSE: {:.5f}'.format(rmse))
```

```python
for name, model in models.items():
    y_pred = model.predict(X_test)
    r2 = get_r2(y_test, y_pred)
    print(name + ' R^2: {:.5f}'.format(r2))
```

```python
# Create a linear regression model
model3 = LinearRegression()

# Train the model on the training data
model3.fit(X_train, y_train)

# Predict target values for the test data
pred3 = model3.predict(X_test)

# Evaluate the model's performance
print("R2 score: ", model3.score(X_test, y_test))
```
